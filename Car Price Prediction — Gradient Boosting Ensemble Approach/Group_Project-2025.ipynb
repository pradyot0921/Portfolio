### 1. Forecasting problem 

The forecasting problem in this competition is to predict the listed price of used cars based on specific details of the cars procided in the training and testing datasets, like, dealer details, and the vehicle's features like colour, mileage, etc. Each observation represents an individual car listing, with its inherent features, e.g. make, model, engine type, horsepower, etc. The objective is to build a predictive model that accurately estimates the car’s market price through training on the features of these vehicles.

This is a regression problem where the target variable 'price' is continuous and numeric.

Such predictive models can be applied in automotive and vehicle dealership industries, e.g.:

- Online dealerships can use these models to predict prices of vehicles.
- Individual sellers and buyers can use the models to judge the worth of vehicles based on their features, rather than relying on instincts or market trends.
- Financial institutions can also use such models to predict the pricing of vehicles, basis which micro-loans can be approved for consumers with less risk to the bank by having accurate forcasts.

### 2. Evaluation criteria 

The competition evaluates model performance using the Mean Absolute Percentage Error (MAPE). This metric calculates the percentage of error between the predicted and actual results, and presents an average of their absolute values.

Advantages:

- MAPE is easy to understand as it is unitless and provides a value in percentage, example a MAPE of 0.1 would easily tell us that a model on average is able to predict 90% of the values well, with an average of 10% error 

Drawbacks:

- If a real value is small, and the predicted value is much smaller (e.g. 1 and 0.1 respectively), MAPE will predict this as a very high error. With such cases, the MAPE of the model can appear very high due to some of the values being mis-predicted, while others might actually have a low percentage error.
  
Due to the major drawback above, we chose to evaluate our models through the training and validation processes using the Root Mean Square Error (RMSE), along with the R^2, which are more widely accepted as the industry standards, and are immune to this drawback as RMSE gives us the real value in the mean difference in roots of sum of squares of residuals, which has same units as the target variable, thus giving a clearer picture on how much predicted values may be distant from the actual ones. That combined with R2 shows us exactly how the model fits the data and by what factor are the predicted prices off from the real ones, on average.

```python
# Define a helper function to extract numbers from mixed text
_num_pat = re.compile(r"[-+]?\d*\.?\d+")
def extract_float(x):
    if pd.isna(x):
        return np.nan
    m = _num_pat.search(str(x))
    return float(m.group()) if m else np.nan

# Helper: clean engine_type to keep only codes like 'I4', 'V6', 'H4', etc.
def clean_engine_type(x):
    if pd.isna(x):
        return np.nan
    match = re.search(r'\b[IVH]\d\b', str(x).upper())
    return match.group(0) if match else np.nan
```

```python
# Applying this cleaning function to all string-based numeric columns
# Columns to clean
columns_to_clean = ["back_legroom", "front_legroom", "height", "width", "wheelbase", "maximum_seating"]

# Applying cleaning to both train and test
for df in (train, test):
    for col in columns_to_clean:
        if col in df.columns:
            df[col] = df[col].apply(extract_float)

    # Clean engine_type separately 
    if "engine_type" in df.columns:
        df["engine_type"] = df["engine_type"].apply(clean_engine_type)
```

```python
## Task 2, Question 2 Code Here

# Helper functions (return two columns each)
def parse_power(text):
    if pd.isna(text):
        return pd.Series({"power_hp": np.nan, "power_rpm": np.nan})
    nums = re.findall(r"\d[\d,]*\.?\d*", str(text))
    nums = [float(n.replace(",", "")) for n in nums]
    return pd.Series({"power_hp": nums[0] if len(nums) > 0 else np.nan,
                      "power_rpm": nums[1] if len(nums) > 1 else np.nan})

def parse_torque(text):
    if pd.isna(text):
        return pd.Series({"torque_lbft": np.nan, "torque_rpm": np.nan})
    nums = re.findall(r"\d[\d,]*\.?\d*", str(text))
    nums = [float(n.replace(",", "")) for n in nums]
    return pd.Series({"torque_lbft": nums[0] if len(nums) > 0 else np.nan,
                      "torque_rpm": nums[1] if len(nums) > 1 else np.nan})
```

```python
for df in (train, test):
    # new numeric features from multi-value text
    if "power" in df.columns:
        df[["power_hp", "power_rpm"]] = df["power"].apply(parse_power)
    if "torque" in df.columns:
        df[["torque_lbft", "torque_rpm"]] = df["torque"].apply(parse_torque)

    # Vehicle age derived from year and listing date
    if "listed_date" in df.columns and "year" in df.columns:
        df["listed_date"] = pd.to_datetime(df["listed_date"], errors="coerce")
        df["vehicle_age"] = df["listed_date"].dt.year - df["year"]

    # Hp per Litre derived from horsepower and engine displacement 
    if "horsepower" in df.columns and "engine_displacement" in df.columns:
        df["hp_per_litre"] = df["horsepower"] / df["engine_displacement"]

    # Avg Mileage derived from city fuel economy and highway fuel economy
    if "city_fuel_economy" in df.columns and "highway_fuel_economy" in df.columns:
        df["avg_mileage"] = (df["city_fuel_economy"]+df["highway_fuel_economy"])/2

    if "engine_type" in df.columns:
        df["engine_layout"] = df["engine_type"].str.extract(r"([A-Za-z])")
        df["cylinder_count"] = df["engine_type"].str.extract(r"(\d+)").astype(float)

    # Dropping the original columns
    cols_to_drop = ["power", "torque", "listed_date","engine_type"]
    df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)
```

```python
## Task 2, Question 4 Code Here
# Task 2 — Q4: Encode categorical variables 
# Importing necessary library
from sklearn.preprocessing import OneHotEncoder
```

```python
# Splitting features (exclude target & IDs)
train_X = train.drop(columns=[c for c in [TARGET] + ID_COLS if c in train.columns])
test_X  = test.drop(columns=[c for c in ID_COLS if c in test.columns])
```

```python
# Identifying categorical columns
cat_cols = train_X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()
cat_cols = [c for c in cat_cols if c.lower() not in [id_col.lower() for id_col in ID_COLS]]
```

```python
# Helpers for reduction rules
SEP_PAT = re.compile(r"[\/,;|]")

def is_multi_value(x):
    if pd.isna(x):
        return False
    return bool(SEP_PAT.search(str(x)))

def normalise_token(x):
    if pd.isna(x):
        return x
    return str(x).strip().title()

def reduce_to_top5_plus_other(train_col, test_col):
    # Normalise; multi-value -> "Other"
    tr = train_col.astype("object").map(normalise_token)
    te = test_col.astype("object").map(normalise_token)
    tr = tr.where(~tr.map(is_multi_value), other="Other")
    te = te.where(~te.map(is_multi_value), other="Other")

    # If >5 uniques in TRAIN, keep top-5 + "Other"
    if tr.dropna().nunique() > 5:
        top5 = tr.value_counts(dropna=True).head(5).index.tolist()
        def map_func(v):
            if pd.isna(v):
                return np.nan
            return v if v in top5 else "Other"
        tr = tr.map(map_func)
        te = te.map(map_func)
    return tr, te
```

```python
# Applying reduction to all categorical columns
train_cat = pd.DataFrame(index=train_X.index)
test_cat  = pd.DataFrame(index=test_X.index)

for col in cat_cols:
    tr_m, te_m = reduce_to_top5_plus_other(train_X[col], test_X[col])
    train_cat[col] = tr_m
    test_cat[col]  = te_m
```

```python
# Examining skewness
skew_vals = train_ready[num_cols].apply(lambda x: skew(x.dropna()))
skewed = skew_vals[abs(skew_vals) > 1].sort_values(ascending=False)
print("\nHighly skewed numeric features (|skew| > 1):")
print(skewed)
```

```python
# iterate over column names, not values
skewed_cols = skewed.index.tolist()
```

```python
# Detecting outliers BEFORE clipping
print("\nChecking for outliers using the IQR rule:")
for col in num_cols:
    Q1, Q3 = train_ready[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = ((train_ready[col] < lower) | (train_ready[col] > upper)).sum()
    if outliers > 0:
        print(f"{col}: {outliers} outliers ({outliers / len(train_ready) * 100:.2f}%)")
```

```python
# Reducing strong right-skew with log1p (only for non-negative features)
for col in skewed_cols:
    if (train_ready[col].min() >= 1) and (col in test_ready.columns and test_ready[col].min() >= 1):
        train_ready[col] = np.log1p(train_ready[col])
        test_ready[col]  = np.log1p(test_ready[col])
```

```python
# Clipping outliers using IQR (keeping all rows)
for col in num_cols:
    unique_vals = train_ready[col].dropna().unique()
    if len(unique_vals) <= 2 and set(unique_vals).issubset({0, 1}):
        # Skip binary/dummy columns
        continue

    q1, q3 = train_ready[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    if iqr == 0:  # prevent divide by zero
        continue
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    train_ready[col] = np.clip(train_ready[col], lower, upper)
```

```python
# Importing Necessary Libraries
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, make_scorer
from sklearn.experimental import enable_hist_gradient_boosting 
from sklearn.ensemble import HistGradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from lightgbm import LGBMRegressor, early_stopping, log_evaluation
from joblib import parallel_backend
```

```python
# y from  train_ready or train 
if TARGET in train_ready.columns:
    y = train_ready[TARGET].copy()
else:
    y = train[TARGET].copy()

# X from train_ready: drop target/IDs if present, keep numeric only
X = train_ready.drop(columns=[c for c in [TARGET] + ID_COLS if c in train_ready.columns], errors="ignore")
X = X.select_dtypes(include=["number"]).copy()
```

```python
# Hist Gradient Boost Base model 
hgb_base = HistGradientBoostingRegressor(
    loss="squared_error",
    learning_rate=0.05,        # will be tuned
    max_iter=1000,             # high cap; ES will usually stop earlier
    early_stopping=True,       # enable early stopping
    n_iter_no_change=50,       # patience
    validation_fraction=0.1,   # ES validation split (from training fold)
    random_state=42
)

# Parameter grid 
param_grid = {
    "learning_rate": [0.02, 0.03, 0.05],
    "max_leaf_nodes": [31, 63, 127],
    "min_samples_leaf": [20, 50, 100],
    "l2_regularization": [0.0, 0.1, 1.0],
    "max_bins": [255],
}

# Cross Validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Grid Search 
hgb_grid = GridSearchCV(
    estimator=hgb_base,
    param_grid=param_grid,
    scoring="neg_root_mean_squared_error",  # minimise RMSE
    cv=cv,
    n_jobs=1,    # keep 1 to dodge the Windows wmic/joblib issue
    verbose=1,
    refit=True
)

# Fitting (ES runs inside each CV fold)
with parallel_backend('threading', n_jobs=-1):
    hgb_grid.fit(X_train, y_train)
    
hgb_best = hgb_grid.best_estimator_
cv_rmse = -hgb_grid.best_score_

# Evaluating on holdout/validation split 
y_pred_tr = hgb_best.predict(X_train)
y_pred_va = hgb_best.predict(X_valid)

r2_train = r2_score(y_train, y_pred_tr)
r2_valid = r2_score(y_valid, y_pred_va)
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_tr))
rmse_valid = np.sqrt(mean_squared_error(y_valid, y_pred_va))

# Some sklearn versions expose n_iter_ (best number of boosting iterations); guard it:
best_iter = getattr(hgb_best, "n_iter_", None)

print("\nHistGradientBoosting (Tuned via CV with Early Stopping)")
print("Best Params:", hgb_grid.best_params_)
print(f"CV RMSE (5-fold): {cv_rmse:,.2f}")
print(f"Train → R²: {r2_train:.3f} | RMSE: {rmse_train:,.2f}")
print(f"Valid → R²: {r2_valid:.3f} | RMSE: {rmse_valid:,.2f}")
print(f"Best Iteration (n_iter_): {best_iter}")
```

```python
# Light GBM Model
# GridSearchCV (no leakage) 
lgbm = LGBMRegressor(random_state=42, verbosity=-1)

lgbm_params = {
    "n_estimators": [200, 300],
    "max_depth": [-1, 4, 6],
    "learning_rate": [0.03, 0.05, 0.07],
    "subsample": [0.7, 0.8],
    "colsample_bytree": [0.7, 0.8]
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)
lgbm_grid = GridSearchCV(
    estimator=lgbm,
    param_grid=lgbm_params,
    scoring="neg_root_mean_squared_error",   # minimise RMSE
    cv=cv,
    n_jobs=-1,
    verbose=1,
    refit=True
)
lgbm_grid.fit(X_train, y_train)

lgbm_best = lgbm_grid.best_estimator_
cv_rmse = -lgbm_grid.best_score_

# Holdout (pre-ES) 
y_pred_lgbm_train = lgbm_best.predict(X_train)
y_pred_lgbm_valid = lgbm_best.predict(X_valid)

lgbm_r2_train = r2_score(y_train, y_pred_lgbm_train)
lgbm_r2_valid = r2_score(y_valid, y_pred_lgbm_valid)
lgbm_rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_lgbm_train))
lgbm_rmse_valid = np.sqrt(mean_squared_error(y_valid, y_pred_lgbm_valid))

# Displaying the result
print("\n🔹 LightGBM (Tuned via CV)")
print("Best Parameters:", lgbm_grid.best_params_)
print(f"CV RMSE (5-fold): {cv_rmse:,.2f}")
print(f"Pre-ES  Train → R²: {lgbm_r2_train:.3f} | RMSE: {lgbm_rmse_train:,.2f}")
print(f"Pre-ES  Valid → R²: {lgbm_r2_valid:.3f} | RMSE: {lgbm_rmse_valid:,.2f}")
print("⚠️ Potential overfit." if (lgbm_r2_train - lgbm_r2_valid) > 0.05 else "✅ Generalises reasonably.")

# Early-stopping refinement 
best_params_es = lgbm_grid.best_params_.copy()
best_params_es["n_estimators"] = 10_000  # large cap; ES will stop early

lgbm_es = LGBMRegressor(
    random_state=42,
    verbosity=-1,
    **best_params_es
)

# Fitting with early stopping
lgbm_es.fit(
    X_train, y_train,
    eval_set=[(X_valid, y_valid)],
    eval_metric="rmse",
    callbacks=[
        early_stopping(stopping_rounds=200, first_metric_only=True),
        log_evaluation(0)
    ]
)

best_iter = getattr(lgbm_es, "best_iteration_", None)
if best_iter is None:
    best_iter = lgbm_grid.best_params_.get("n_estimators", 300)

# Post-ES performance
y_pred_lgbm_valid_es = lgbm_es.predict(X_valid, num_iteration=best_iter)
lgbm_r2_valid_es = r2_score(y_valid, y_pred_lgbm_valid_es)
lgbm_rmse_valid_es = np.sqrt(mean_squared_error(y_valid, y_pred_lgbm_valid_es))

# Displaying the Post-ES results
print("\n✨ Early-Stopping Refinement")
print(f"best_iteration_: {best_iter}")
print(f"Post-ES Valid → R²: {lgbm_r2_valid_es:.3f} | RMSE: {lgbm_rmse_valid_es:,.2f}")
```
